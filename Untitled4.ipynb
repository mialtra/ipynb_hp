```python
import pandas as pd
import numpy as np
from pmdarima.arima import auto_arima
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

from datetime import datetime,date, timedelta

# Load data historis saham Telkom
KODE = 'TLKM.JK'
today_date = date.today()
td = timedelta(366)
mulai = today_date - td
end_ = today_date + timedelta(1)

import yfinance as yf

df = yf.download(KODE, start=mulai, end = end_)

# Pra-pemrosesan data
#df['Date'] = pd.to_datetime(df['Date'])
#df = df.set_index('Date')
df = df[['Close']]
train_data = df[:int(len(df)*0.8)]
test_data = df[int(len(df)*0.8):]

# Lakukan prediksi dengan Auto-ARIMA
model = auto_arima(train_data, trace=True, error_action='ignore', suppress_warnings=True)
model.fit(train_data)

# Prediksi harga saham Telkom di masa depan
forecast = model.predict(n_periods=len(test_data))
forecast = pd.DataFrame(forecast, index=test_data.index, columns=['Prediction'])

# Hitung nilai RMSE
mse = mean_squared_error(test_data, forecast)
rmse = np.sqrt(mse)
print('RMSE:', rmse)

# Plot data dan prediksi
plt.figure(figsize=(16,8))
plt.plot(train_data, label='Training Data')
plt.plot(test_data, label='Test Data')
plt.plot(forecast, label='Predicted Data')
plt.title('Saham Telkom Prediction with Auto-ARIMA')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.show()
```
